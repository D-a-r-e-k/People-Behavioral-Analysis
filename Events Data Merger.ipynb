```python
import os

files = os.listdir('Data')

print(f"Total files: {len(files)}")
      
files = list(set(files))

print(f"Total unique files: {len(files)}")
```

```python
files = sorted(files)
```

```python
from tqdm import tqdm
import pandas as pd
```

```python
stations_with_labels = pd.read_csv("stations_with_labels.csv").loc[:,["description","serial", "is_outdoor", "is_road_nearby_values", "is_city_center", "is_transportation_node","is_park","is_office_nearby", "is_recreation_zone"]]

stations_with_labels
```

```python
d1 = pd.read_csv(f"Data/{files[0]}")
d2 = pd.read_csv(f"Data/{files[1]}")

d1.merge(stations_with_labels, on="serial")

d1.append(d2).reset_index(drop=True)
```

```python
result = None
for f in tqdm(files):
    content = pd.read_csv(f"Data/{f}")[["time","hash","serial","distance"]]
    #content_with_stations = content.merge(stations_with_labels, on="serial")
    
    if result is None:
        result = content
    else:
        result = result.append(content)
        
result = result.reset_index(drop=True)
```

```python
result.to_csv("tracking_data_all.csv",index=False)
```
